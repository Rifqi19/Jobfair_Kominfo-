Persentase status vaksin yang disyaratkan perusahaan.

```python
# Contoh data status vaksin pegawai (0, 1, atau 2) dari kolom 'vaccine_status'
status_pegawai = job_vacancies['vaccine_status'].value_counts()

# Menghitung jumlah setiap status vaksin
jumlah_belum_vaksin = status_pegawai.get(0, 0)
jumlah_vaksin_dosis_1 = status_pegawai.get(1, 0)
jumlah_vaksin_dosis_2 = status_pegawai.get(2, 0)

# Menghitung total pegawai
total_pegawai = jumlah_belum_vaksin + jumlah_vaksin_dosis_1 + jumlah_vaksin_dosis_2

# Menghitung persentase setiap status vaksin
persentase_belum_vaksin = (jumlah_belum_vaksin / total_pegawai) * 100
persentase_vaksin_dosis_1 = (jumlah_vaksin_dosis_1 / total_pegawai) * 100
persentase_vaksin_dosis_2 = (jumlah_vaksin_dosis_2 / total_pegawai) * 100

# Menampilkan hasil
print(f"Jumlah BELUM VAKSIN: {jumlah_belum_vaksin}, Persentase: {persentase_belum_vaksin:.2f}%")
print(f"Jumlah VAKSIN DOSIS 1: {jumlah_vaksin_dosis_1}, Persentase: {persentase_vaksin_dosis_1:.2f}%")
print(f"Jumlah VAKSIN DOSIS 2: {jumlah_vaksin_dosis_2}, Persentase: {persentase_vaksin_dosis_2:.2f}%")
```

Jumlah jam kerja yang ditawarkan perusahaan.

```python
# Menghitung total jumlah jam kerja yang ditawarkan perusahaan
total_jam_kerja = job_vacancies['number_of_hours_per_week'].sum()

# Menampilkan hasil
print(f"Total jam kerja yang ditawarkan perusahaan: {total_jam_kerja} jam per minggu")
```

```python
# Misalkan job_vacancies adalah DataFrame yang berisi kolom 'number_of_hours_per_week'
# Jika belum terdapat DataFrame, pastikan untuk membaca atau menginisialisasi datanya terlebih dahulu

# Menentukan batas kategori jam kerja
batas_kategori = [0, 40, 80, 120, float('inf')]

# Membuat kategori jam kerja menggunakan pd.cut()
kategori_jam_kerja = pd.cut(job_vacancies['number_of_hours_per_week'], bins=batas_kategori, labels=['<=40', '41-80', '81-120', '>120'])

# Menghitung jumlah jam kerja dalam setiap kategori
jumlah_per_kategori = kategori_jam_kerja.value_counts().sort_index()

# Menampilkan hasil
print("Jumlah jam kerja dalam setiap kategori:")
print(jumlah_per_kategori)

# Menghitung total jam kerja dari semua kategori
total_jam_kerja = jumlah_per_kategori.sum()

# Menampilkan total jam kerja
print(f"Total jam kerja: {total_jam_kerja}")
```
