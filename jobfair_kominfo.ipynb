```python
import pandas as pd
```

```python
data_jobfair = pd.read_excel('data_jobfair.xlsx')
data_jobfair
```

```python
data_jobfair.isna().sum()

# Langkah 1: Hapus baris dengan data kosong di kolom "Date of Birth"
data_jobfair = data_jobfair.dropna(subset=["date_of_birth"])

# Langkah 2: Ganti data kosong di kolom lain dengan "-"
data_jobfair= data_jobfair.fillna("-")
data_jobfair
```

Usia saat melamar kerja

```python
data_jobfair['usia_saat_melamar'] = (pd.to_datetime(data_jobfair['updatedAt']) - pd.to_datetime(data_jobfair['date_of_birth'])).astype('<m8[Y]')
data_jobfair
```

```python
# Assuming data_jobfair is a Pandas DataFrame with the 'usia_saat_melamar' column created previously

# Find the index of rows where 'usia_saat_melamar' is less than 15
index_to_drop = data_jobfair[data_jobfair['usia_saat_melamar'] < 15].index

# Drop the rows with the specified index
data_jobfair = data_jobfair.drop(index_to_drop)
```
