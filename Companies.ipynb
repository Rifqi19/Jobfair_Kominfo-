```python
file_path = 'data_jobfair.xlsx'
dataframes = pd.read_excel(file_path, sheet_name=None)
dataframes2 =pd.read_excel(file_path, sheet_name=None)
dataframes3 =pd.read_excel(file_path, sheet_name=None)
```

```python
job_vacancies = dataframes['jobvacancies']
companies = dataframes2['companies']
masters = dataframes3['masters']
masters = masters[masters['category'] == 'company_category']
```

Proporsi kategori perusahaan yang terdata (id category dapat dicocokkan dengan yang tertera di sheet masters)

```python
# Merge the 'companies' DataFrame with the 'masters' DataFrame based on 'category'
merged_data = pd.merge(companies, masters, how='left', left_on='category', right_on='category')

# Calculate the proportion of each category
category_proportions = merged_data['label'].value_counts(normalize=True)

# Display the result
print(category_proportions)
```
